```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
breast_cancer_data = load_breast_cancer()
target = breast_cancer_data["target"]
breast_cancer_data = pd.DataFrame(breast_cancer_data["data"],columns=breast_cancer_data["feature_names"])
breast_cancer_data["target"] = target
breast_cancer_data.head()
```

```python
breast_cancer_data.info()
```

```python
sns.scatterplot(data=breast_cancer_data,y='mean radius',x="texture error")
```

```python
breast_cancer_data.corr()
```

```python
breast_cancer_data.describe()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaling=StandardScaler()
breast_cancer_data =  scaling.fit_transform(breast_cancer_data.drop("target",axis=1))
```

```python
pca = PCA(n_components=3)
pca.fit(breast_cancer_data)
```

```python
pd.DataFrame(pca.components_)
```

```python
sns.heatmap(pca.components_)
```

```python
pca.explained_variance_
```

```python
data = pca.transform(breast_cancer_data)
```

```python
pca_Data = pd.DataFrame(data,columns=["PCA1","PCA2","PCA3"])
```

```python
import plotly.express as px
fig = px.scatter_3d(pca_Data ,x="PCA1", y="PCA2", z="PCA3",
              )
fig.show()
```
