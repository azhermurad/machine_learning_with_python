# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix,ConfusionMatrixDisplay
from datetime import datetime
```

```python
def date_time_to_seconds(date_str):
    # Parse the date-time string into a datetime object
    date_time_obj = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
    # Convert to seconds since the Unix epoch
    epoch = datetime(1970, 1, 1)
    total_seconds = int((date_time_obj - epoch).total_seconds())

    return total_seconds

ad_data["Timestamp"] = ad_data["Timestamp"].apply(date_time_to_seconds)
X = ad_data.drop(["Ad Topic Line","City","Country","Clicked on Ad",'Timestamp'],axis=1)
y = ad_data["Clicked on Ad"]

```

```python
# now we have to split the data to into text and training 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

** Train and fit a logistic regression model on the training set.**

```python
model = LogisticRegression(max_iter=1000)  
model.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
# Make predictions
y_pred = model.predict(X_test)
```

** Create a classification report for the model.**

```python
print(classification_report(y_test, y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python

# Optional: Display the confusion matrix visually
ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
```
